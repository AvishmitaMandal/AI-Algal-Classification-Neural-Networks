```python
#Without normalization

import pandas as pd
import cv2
import numpy as np

class_values = top_5_filtered_df['library_id'].unique()
sampled_images_woNorm = []

for class_value in class_values:
    class_rows = top_5_filtered_df[top_5_filtered_df['library_id'] == class_value]
    images = class_rows['image'].tolist()

    for image_path in images:
        image = cv2.imread(image_path)
        if image is not None:
            image = cv2.resize(image, (224, 224))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            sampled_images_woNorm.append({"image": image, "class_label": class_value})
        else:
            print(f"Error reading image: {image_path}")

```

```python
import pandas as pd
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = np.array([sampled_image["image"] for sampled_image in sampled_images_woNorm])
y = np.array([sampled_image["class_label"] for sampled_image in sampled_images_woNorm])

label_encoder = LabelEncoder()

# Fit LabelEncoder to your class labels and transform them
y_encoded = label_encoder.fit_transform(y)

print("Class labels:", np.unique(y))
# Print the unique encoded class labels
print("Encoded class labels:", np.unique(y_encoded))

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
num_classes = len(np.unique(y_train))

print("Number of unique class labels in training data:", num_classes)
print("Unique class labels in training data:", np.unique(y_train))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Define the CNN model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(54, activation='softmax')  # Adjust num_classes based on your dataset
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc}')

# Plotting the training and validation accuracy over epochs
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()

# Plotting the training and validation loss over epochs
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
import pandas as pd
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Initialize an ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.4, random_state=42)

# Apply data augmentation only on the training set
X_train_augmented = []
y_train_augmented = []

for image, label in zip(X_train, y_train):
    # Reshape image to fit the datagen requirements (4D tensor)
    image = image.reshape((1,) + image.shape)
    # Generate augmented images
    for batch in datagen.flow(image, batch_size=1):
        augmented_image = batch[0].astype('uint8')
        X_train_augmented.append(augmented_image)
        y_train_augmented.append(label)
        break  # break the loop to generate only one augmented image per original image

# Convert augmented data lists to arrays
X_train_augmented = np.array(X_train_augmented)
y_train_augmented = np.array(y_train_augmented)

# Concatenate the original training data with the augmented training data
X_train_final = np.concatenate((X_train, X_train_augmented))
y_train_final = np.concatenate((y_train, y_train_augmented))

# Reshape the training and testing data to remove the extra dimension
X_test = X_test.reshape((-1, 100, 100, 3))
X_train = X_train_final.reshape((-1, 100, 100, 3))

# Optionally, encode the class labels if needed
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit LabelEncoder to your class labels and transform them
y_train = label_encoder.fit_transform(y_train_final)
y_test = label_encoder.transform(y_test)

print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import plot_model

# Define the early stopping callback
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss',  # Monitor validation loss
    patience=3,           # Number of epochs with no improvement after which training will be stopped
    restore_best_weights=True  # Restore the best model weights
)

# Define the learning rate schedule
def learning_rate_schedule(epoch):
    if epoch < 5:
        return 0.001  # Use a high learning rate for the initial epochs
    elif epoch < 10:
        return 0.0005  # Decrease the learning rate after a few epochs
    else:
        return 0.0001  # Further decrease the learning rate after more epochs

lr_scheduler = callbacks.LearningRateScheduler(learning_rate_schedule)
num_classes = 10

# Define the CNN model with regularization (dropout)
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dropout(0.5),  # Add dropout layer with dropout rate of 0.5
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')  # Adjust num_classes based on your dataset
])

# Plot the model architecture
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)

# Compile the model with adjusted learning rate and different optimizer parameters
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)  # Adjust initial learning rate as needed
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model with early stopping and learning rate schedule
history = model.fit(X_train, y_train, epochs=40, batch_size=64,
                    validation_split=0.2, callbacks=[early_stopping, lr_scheduler])

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

# Function to create a collage from a list of images
def create_collage_with_predictions(images, true_labels, predicted_labels, predicted_probs, top_classes, class_labels, selected_indices, collage_size):
    width, height = images[0].size
    collage = Image.new('RGB', (width * collage_size[1], height * collage_size[0]))
    draw = ImageDraw.Draw(collage)
    for i in range(len(images)):
        row = i // collage_size[1]
        col = i % collage_size[1]
        image = images[i]
        true_label = true_labels[i]
        predicted_label = predicted_labels[i]
        predicted_prob = predicted_probs[selected_indices[i]]
        top_classes_idx = top_classes[selected_indices[i]]

        # Paste image onto collage
        collage.paste(image, (col * width, row * height))

        # Add bounding box with true class label
        draw.rectangle([((col * width) + 5, (row * height) + 5), ((col * width) + 60, (row * height) + 20)], outline="white", width=1)
        draw.text(((col * width) + 5, (row * height) + 5), f'Actual: {true_label}', fill="white")

        # Add predicted labels and their probabilities
        for j, class_idx in enumerate(top_classes_idx):
            class_label = class_labels[class_idx]
            prob = predicted_prob[j]
            draw.text(((col * width) + 5, (row * height) + 20 + (j * 15)), f'{class_label}: {prob:.2f}', fill="white")

    return collage

# Make predictions on the test set
predictions = model.predict(X_test)

# Get the top 3 predicted classes and their probabilities for each image
top_classes = np.argsort(predictions, axis=1)[:, ::-1][:, :3]
predicted_probs = np.take_along_axis(predictions, top_classes, axis=1)

# Select a few images to display
num_images_to_display = 12
selected_indices = np.random.choice(len(X_test), num_images_to_display, replace=False)

# print(len(X_test))
# print(len(y_test))
# print(len(top_classes))

# Collect the selected images, true labels, and predicted labels
selected_images = [Image.fromarray(X_test[idx]) for idx in selected_indices]
true_labels = y_test[selected_indices]
predicted_labels = top_classes[selected_indices]

# Create the collage with predicted labels
collage_with_predictions = create_collage_with_predictions(selected_images, true_labels, predicted_labels, predicted_probs, top_classes, class_labels, selected_indices, (3, 4))

# Display the collage with predicted labels
plt.figure(figsize=(8, 6))
plt.imshow(collage_with_predictions)
plt.axis('off')
plt.show()
```
